   ## 一、寻医问药网医疗信息爬取

```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
```

  ### 1. 爬取按科室查询的科室url信息

```python
org_url = 'http://3g.jib.xywy.com/?fromurl=3ghome'
org_response = requests.get(org_url)
org_soup = BeautifulSoup(org_response.content,'html.parser')
# links = soup.select('li',{'class':'symptoms__container--item'})
sec_links =org_soup.select('body>div.sec-para.depart-check>div>ul>li>a')

sec_urls = []
for link in sec_links:
    text = link.text
    sec_url = link['href']
    sec_urls.append(link['href'])
print(sec_urls)
```

   ### 2. 创建需要获取的信息的空列表

```python
Name = []
Is_common = []
Else_name = []
Overview = []
Department = []
RegularTest = []
Treatment = []
Drug = []
Cost = []
Contagious = []
Healing_cycle = []
Healing_rate = []
Prevalence_rate = []
Prevalent_population = []
Related_symptoms = []
Related_diseases = []
Symptom_analysis = []
Causes_analysis = []
Diagnosis_analysis = []
Treatment_analysis = []
Health_Care_analysis = []
Prevention_analysis = []
Examination_analysis = []
Complications_analysis = []
```

   ### 3. 逐页爬取疾病相关信息并将其储存在列表中

```python
for i in range(1,10140):
    dis_url = 'http://3g.jib.xywy.com/il_sii_'+str(i)+ '.html'
    dis_response = requests.get(dis_url)
    if dis_response.status_code == 200:
        dis_soup = BeautifulSoup(dis_response.content,'html.parser')

        #疾病名称
        dis_name = dis_soup.select_one('em')
        if dis_name != None:
            Name.append(dis_name.text)
        else:
            Name.append('None')

        #是否是常见病
        dis_is_common = dis_soup.select_one('div.Disease-title.clearfix>span')
        if dis_is_common != None:
            Is_common.append(dis_is_common.text)
        else:
            Is_common.append('None')

        #别名
        dis_else_name = dis_soup.select_one('div.Disease-alias')
        if dis_else_name != None:
            Else_name.append(dis_else_name.text.replace('别名：',''))
        else:
            Else_name.append('None')

        #概述
        dis_overview = dis_soup.select_one('div.Disease-brief')
        if dis_overview != None:
            Overview.append(dis_overview.text.strip())
        else:
            Overview.append('None')

        dis_info = dis_soup.select('div.reach-right.fl')
        dis_info_tab = dis_soup.select('div.reach-tabright.fl')

        if dis_info[0] != None:
            Department.append(dis_info[0].text.replace('\u2003',' ').strip())
        else:
            Department.append('None')

        #需做检查
        if dis_info[1] != None:
            RegularTest.append(dis_info[1].text.replace('\xa0','/').replace(' ','').strip())
        else:
            RegularTest.append('None')

        #治疗方法
        if dis_info[2] != None:
            Treatment.append(dis_info[2].text.replace('\xa0','/').replace(' ','').strip())
        else:
            Treatment.append('None')

        #常用药物
        if dis_info_tab[0] != None:
            Drug.append(dis_info_tab[0].text.replace(' ','').replace('\n',' ').strip())
        else:
            Drug.append('None')    

        #一般费用
        if dis_info[3] != None:
            Cost.append(dis_info[3].text.replace('\xa0','/').replace(' ','').strip())
        else:
            Cost.append('None')   

        #传染性
        if dis_info[4] != None:
            Contagious.append(dis_info[4].text.replace('\xa0','/').replace(' ','').strip())
        else:
            Contagious.append('None')       

        #治愈周期
        if dis_info[5] != None:
            Healing_cycle.append(dis_info[5].text.replace('\xa0','/').replace(' ','').strip())
        else:
            Healing_cycle.append('None')          

        #治愈率
        if dis_info[6] != None:
            Healing_rate.append(dis_info[6].text.replace('\xa0','/').replace(' ','').strip())
        else:
            Healing_rate.append('None')       

        #患病比率
        if dis_info[7] != None:
            Prevalence_rate.append(dis_info[7].text.replace('\xa0','/').replace(' ','').strip())
        else:
            Prevalence_rate.append('None')      

        #好发人群
        if dis_info[8] != None:
            Prevalent_population.append(dis_info[8].text.replace('\xa0','/').replace(' ','').strip())
        else:
            Prevalent_population.append('None')       

        #相关症状
        if dis_info_tab[1] != None:
            Related_symptoms.append(dis_info_tab[1].text.replace(' ','').replace('\n',' ').strip())
        else:
            Related_symptoms.append('None') 

        #相关疾病
        if dis_info_tab[2] != None:
            Related_diseases.append(dis_info_tab[2].text.replace(' ','').replace('\n',' ').strip())
        else:
            Related_diseases.append('None')


        dis_detail = dis_soup.select('div.analysis-nei')

        #症状解析
        if dis_detail[0] != None:
            Symptom_analysis.append(dis_detail[0].text.replace('\n\r\n\t',' ').strip())
        else:
            Symptom_analysis.append('None')   


        #病因解析
        if dis_detail[1] != None:
            Causes_analysis.append(dis_detail[1].text.replace('\n\r\n\t',' ').strip())
        else:
            Causes_analysis.append('None')   

        #诊断解析
        if dis_detail[2] != None:
            Diagnosis_analysis.append(dis_detail[2].text.replace('\n\r\n\t',' ').strip())
        else:
            Diagnosis_analysis.append('None') 

        #治疗解析
        if dis_detail[3] != None:
            Treatment_analysis.append(dis_detail[3].text.replace('\n\r\n\t',' ').strip())
        else:
            Treatment_analysis.append('None') 

        #饮食保健
        if dis_detail[4] != None:
            Health_Care_analysis.append(dis_detail[4].text.replace('\n\r\n\t',' ').strip())
        else:
            Health_Care_analysis.append('None') 

        #预防解析
        if dis_detail[5] != None:
            Prevention_analysis.append(dis_detail[5].text.replace('\n\r\n\t',' ').strip())
        else:
            Prevention_analysis.append('None') 

        #检查解析
        if dis_detail[6] != None:
            Examination_analysis.append(dis_detail[6].text.replace('\n\r\n\t',' ').strip())
        else:
            Examination_analysis.append('None') 

        #并发症解析
        if dis_detail[7] != None:
            Complications_analysis.append(dis_detail[7].text.replace('\n\r\n\t',' ').strip())
        else:
            Complications_analysis.append('None') 
    else:
        print('页面不存在或无法访问')
```

   ### 4. 将数据导出到excel文件中

```python
output_excel = {'疾病名称': [], '是否是常见病': [],'别名': [],'概述': [],'科室': [],'需做检查': [],
                '治疗方法': [],'常用药物': [],'一般费用': [],'传染性': [],'治愈周期': [],'治愈率': [],
                '患病比率': [],'好发人群': [],'相关症状': [],'相关疾病': [],'症状解析': [],'病因解析': [],
                '诊断解析': [],'治疗解析': [],'饮食保健': [],'预防解析': [],'检查解析': [],'并发症解析': []}
output_excel['疾病名称'] = Name
output_excel['是否是常见病'] = Is_common
output_excel['别名'] = Else_name
output_excel['概述'] = Overview
output_excel['科室'] = Department
output_excel['需做检查'] = RegularTest
output_excel['治疗方法'] = Treatment
output_excel['常用药物'] = Drug
output_excel['一般费用'] = Cost
output_excel['传染性'] = Contagious
output_excel['治愈周期'] =Healing_cycle
output_excel['治愈率'] = Healing_rate
output_excel['患病比率'] = Prevalence_rate
output_excel['好发人群'] = Prevalent_population
output_excel['相关症状'] = Related_symptoms
output_excel['相关疾病'] = Related_diseases
output_excel['症状解析'] = Symptom_analysis
output_excel['病因解析'] =  Causes_analysis
output_excel['诊断解析'] = Diagnosis_analysis
output_excel['治疗解析'] = Treatment_analysis
output_excel['饮食保健'] = Health_Care_analysis
output_excel['预防解析'] = Prevention_analysis
output_excel['检查解析'] = Examination_analysis
output_excel['并发症解析'] = Complications_analysis


output = pd.DataFrame(output_excel)
output.to_excel('diseases.xlsx',encoding = 'utf-8',index = False)
```
